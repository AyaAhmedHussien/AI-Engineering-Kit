````python
import requests
from transformers import pipeline


classifier = pipeline("sentiment-analysis")
print(classifier("I am happy to be an AI Engineer!"))
````

````python
def analyze_text_sentiment(text):
    """
    Analyzes the sentiment of a given input text using the pre-loaded pipeline.
    Returns a formatted string with the label and confidence score.
    """
    result = classifier(text)
    label = result[0]['label']
    score = result[0]['score']
    return f"Text: '{text}' | Sentiment: {label} | Confidence: {score:.2%}"

# Testing the function
test_input = "Artificial Intelligence is challenging but exciting!"
print(analyze_text_sentiment(test_input))
````

````python
import requests


from google.colab import userdata
api_key = userdata.get('GNEWS_API_KEY')

url = "https://gnews.io/api/v4/search"
params = {
    'q': 'maritime navigation',
    'lang': 'en',
    'token': api_key
}

response = requests.get(url, params=params)

if response.status_code == 200:
    articles = response.json().get('articles', [])
    for article in articles[:3]:
        print(f"Title: {article['title']}")

else:
    print(f"Error: {response.status_code}")
````
